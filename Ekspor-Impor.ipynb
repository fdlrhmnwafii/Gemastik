```python
fig, ax = plt.subplots(2,1)

sns.regplot(x='y_float',y='impor',data=df, ax=ax[0],scatter_kws={'alpha':0.4})
sns.regplot(x='y_float',y='ekspor',data=df, ax=ax[1],scatter_kws={'alpha':0.4})

ax[0].set_title('Regplot Impor')
ax[1].set_title('Regplot Ekspor')
ax[0].set_xlabel('tahun')
ax[1].set_xlabel('tahun')

plt.tight_layout()
```

```python
#Predict Impor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

y1 = df['impor'].values
x = df['y_float'].values.reshape(-1,1)

lm = LinearRegression()
lm.fit(x, y1)

lm.predict(np.array([2024]).reshape(-1,1))
```

```python
#Predict Ekspor
y2 = df['ekspor'].values
x = df['y_float'].values.reshape(-1,1)

lm = LinearRegression()
lm.fit(x, y2)

lm.predict(np.array([2024]).reshape(-1,1))
```

# Evaluation

```python
def get_mse(df, feature, target):
    # Get x, y to model
    y = df[target].values
    x = df[feature].values.reshape(-1,1)
    print('{} ~ {}'.format(target, feature))
    
    # Build and fit the model
    lm = LinearRegression()
    lm.fit(x, y)
    msg = 'model: y = {:.3f} + {:.3f}x'\
            .format(lm.intercept_, lm.coef_[0])
    print(msg)
    
    # Predict and determine MSE
    y_pred = lm.predict(x)
    error = mean_squared_error(y, y_pred)
    print('mse = {:.2f}'.format(error))
    print()
    
get_mse(df,feature='y_float',target='ekspor')
get_mse(df,feature='y_float',target='impor')
```
